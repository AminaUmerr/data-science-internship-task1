# Task 1 — Exploring and Visualizing the Iris Dataset

## Objective

The objective of this task is to explore and visualize the Iris dataset using Python libraries such as pandas, matplotlib, and seaborn.

## Tools Used

- Python
- pandas
- matplotlib
- seaborn


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

print("Libraries imported successfully")
```

## Loading the Dataset


```python
iris = sns.load_dataset("iris")
```

## Dataset Structure and Overview


```python
print("Dataset Shape:")
print(iris.shape)

print("\nColumn Names:")
print(iris.columns)

print("\nFirst 5 Rows:")
print(iris.head())
```

## Checking Missing Values


```python
print(iris.isnull().sum())
```

## Scatter Plot

This plot shows the relationship between sepal length and petal length.


```python
sns.scatterplot(
    x="sepal_length",
    y="petal_length",
    hue="species",
    data=iris
)

plt.title("Sepal Length vs Petal Length")
plt.show()
```

## Histogram

This histogram shows the distribution of sepal length values.


```python
plt.hist(iris["sepal_length"])

plt.title("Distribution of Sepal Length")

plt.xlabel("Sepal Length")

plt.ylabel("Frequency")

plt.show()
```

## Box Plot

The box plot helps identify outliers and spread of values.


```python
sns.boxplot(x=iris["sepal_width"])

plt.title("Box Plot of Sepal Width")

plt.show()
```

# Conclusion

- The Iris dataset contains flower measurements for three species.
- Petal length and sepal length show clear relationships.
- Some variation and possible outliers exist in sepal width.
- Data visualization helps better understand dataset patterns.
